```python
# multitask
import multiprocessing as mp
import datetime
p = mp.Pool(8)

def run_model1(i):
    m = hddm.HDDM(df)
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model2(i):
    m = hddm.HDDM(df,p_outlier=0.05)
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model3(i):
    m = hddm.HDDM(df,include=('sv'))
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model4(i):
    m = hddm.HDDM(df,include=('z'))
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
```

```python
start_t = datetime.datetime.now()

msrs = [
    p.apply_async(run_model1, [1]),
    p.apply_async(run_model2, [2]),
    p.apply_async(run_model3, [3]),
    p.apply_async(run_model4, [4])
    ]
rss = [res.get() for res in msrs]

end_t = datetime.datetime.now()
elapsed_sec = (end_t - start_t).total_seconds()
print("joblib用时: " + "{:.2f}".format(elapsed_sec) + " 秒")
```

```python
# 封装的函数
def parallel(func, *args, show=False, thread=False, **kwargs):
  """
  并行计算
  :param func: 函数，必选参数
  :param args: list/tuple/iterable,1个或多个函数的动态参数，必选参数
  :param show:bool,默认False,是否显示计算进度
  :param thread:bool,默认False,是否为多线程
  :param kwargs:1个或多个函数的静态参数，key-word的字典形式
  :return:返回一个list,包含了与函数动态参数等量的模型数量。
  """
  import time
  from functools import partial
  from pathos.pools import ProcessPool, ThreadPool
  from tqdm import tqdm
  # 冻结静态参数
  p_func = partial(func, **kwargs)
  # 打开进程/线程池
  pool = ThreadPool() if thread else ProcessPool()
  try:
      if show:
          start = time.time()
          # imap方法
          with tqdm(total=len(args[0]), desc="计算进度") as t:  # 进度条设置
              r = []
              for i in pool.imap(p_func, *args):
                  r.append(i)
                  t.set_postfix({'并行函数': func.__name__, "计算花销": "%ds" % (time.time() - start)})
                  t.update()
      else:
          # map方法
          r = pool.map(p_func, *args)
      return r
  except Exception as e:
      print(e)
  finally:
      # 关闭池
      pool.close()  # close the pool to any new jobs
      pool.join()  # cleanup the closed worker processes
      pool.clear()  # Remove server with matching state
```

```python
# 多任务并行计算
def run_ms(funcs,paras):
  """ run multiple models with async processing
    It is necessary to rename sampler like "m.sample(5000,2000,dbname="name1",db="pickle")"

    para: funcs are hddm models
    para: paras are hddm name, data, conditons etc.
  """
  import multiprocessing as mp
  try:
    p = mp.Pool()
    multi_res = [p.apply_async(j[0],j[1]) for j in zip(funcs,paras)]
    rs = [res.get() for res in multi_res]
    return rs
  except Exception as e:
    print(e)
```

```python
#实例
def run_model1(i):
    m = hddm.HDDM(df)
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model2(i):
    m = hddm.HDDM(df,p_outlier=0.05)
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model3(i):
    m = hddm.HDDM(df,include=('sv'))
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m
def run_model4(i):
    m = hddm.HDDM(df,include=('z'))
    m.find_starting_values()
    m.sample(500,200,db="pickle",dbname=str(i))
    return m

rs = run_ms([run_model1,run_model2,run_model3,run_model4],[1,2,3,4])
```
